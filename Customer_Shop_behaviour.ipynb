````python
# importing libraries
import pandas as pd
````

````python
# importing dataset
df=pd.read_csv('customer_shopping_behavior.csv')
df.head()
````

````python
# find the structure of the dataset
df.info()
````

````python
# understanding the statistical features of the numerical data
df.describe()
````

````python
#statistical values for the object dataype
df.describe(include='object')
````

````python
# checking the missing values
df.isnull().sum()
````

````python
# impute the missing value with median reveiw rating within each category
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
````

````python
# Again recheck for null values
df.isnull().sum()
````

````python
# Modifying the heading texts in the columns to Lower case and replacing the spaces () with (_)
df.columns = df.columns.str.lower().str.replace(' ', '_')
````

````python
# Check for the modification applied to the column texts
df.columns
````

````python
# Replace purchase_amount_(usd) to purchase_amount
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
````

````python
# Check for the renaming
df.columns
````

````python
# Create a column for Age Group
labels=('Young Adult','Adult','Middle-Age','Senior')
df['age_group'] = pd.qcut(df['age'],q=4,labels=labels)
````

````python
# Display column for both Age and Age group respectively
df[['age','age_group']].head(30)
````

````python
# create column purchase frequency days
frequency_mapping= { 'Fortnightly':14,'Weekly':7,'Bi-Weekly':14,'Monthly':30,'Quarterly':90,'Annually':365, 'Every 3 Months': 90}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
````

````python
# Check the changes in the dataset
df[['frequency_of_purchases','purchase_frequency_days']].head()
````

````python
# Display for Discount applied and promo code used
df[['discount_applied', 'promo_code_used']].head(10)
````

````python
# Check for Discount applied and promo code used if same or not?
(df['discount_applied'] == df['promo_code_used']).all()
````

````python
# Since both applied are same, so we will drop the column 'promo_code_used'
df= df.drop('promo_code_used',axis=1)
````

````python
# check whether column dropped or not
df.columns
````
